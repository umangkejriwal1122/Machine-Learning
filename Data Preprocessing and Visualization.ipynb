## Data Wrangling

```python
titanic_data.info()
```

```python
mean = np.round(titanic_data['Age'].mean(),2)
```

```python
mean
```

```python
titanic_data['Age'].fillna(mean,inplace=True)
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data = titanic_data.dropna()
```

```python
pd.get_dummies(titanic_data['Sex'])
```

```python
sex = pd.get_dummies(titanic_data['Sex'],drop_first=True)
sex.head()
```

```python
pclass = pd.get_dummies(titanic_data['Pclass'],drop_first=True)
pclass.head()
```

```python
titanic_data.head()
```

```python
titanic_data.drop(["PassengerId","Name","Sex","Ticket","Embarked","Fare"],axis=1,inplace=True)
```

```python
titanic_data.head()
```

```python
titanic_data = pd.concat([titanic_data,sex,pclass],axis=1)
titanic_data.head()
```

```python
titanic_data.drop('Pclass',axis=1,inplace=True)
```

```python
titanic_data.head()
```

```python
titanic_data.rename(columns={'male':"Sex",2:"Parch1",3:"Parch2"},inplace=True)
```
