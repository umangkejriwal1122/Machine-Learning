```python
import pandas as pd
```

```python
emp = pd.read_csv("/content/employees.csv")
emp.head()
```

```python
emp.sort_values("First Name",ascending=True,inplace=True)
emp.head()
```

```python
#### display all duplicate names
mask = emp['First Name'].duplicated()
emp[mask]
```

```python
#### display all unique names
### telda function
mask = ~ emp['First Name'].duplicated()
emp[mask]
```

```python
### Drop duplicates
```

```python
emp.drop_duplicates()   ## compare complete columns
```

```python
emp.drop_duplicates(subset=["First Name"])
```

```python
emp.drop_duplicates(subset=["First Name"],keep="first")  ## consider the first data as unique
```

```python
emp.drop_duplicates(subset=["First Name"],keep="last")  ## consider the last data as unique
```
