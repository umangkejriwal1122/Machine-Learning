```python
import seaborn as sns
import pandas as pd
```

```python
data = sns.load_dataset('titanic')
data.head()
```

```python
data['sex'].value_counts()
```

#### One Hot Encoding 

```python
gender = pd.get_dummies(data['sex'])
gender
```

```python
gender = pd.get_dummies(data['sex'],drop_first=True)
gender
```

```python
pd.concat([data,gender],axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
```

```python
data['embarked'].value_counts()
```

```python
emb = ohe.fit_transform(data[['embarked']])
emb.toarray()
```

```python
emb_df = pd.DataFrame(emb.toarray())
```

```python
emb_df = pd.concat([data,emb_df],axis=1)
```
